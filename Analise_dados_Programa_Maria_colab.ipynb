### Valores discrepantes (outliers)

```python
import matplotlib.pyplot as plt
```

```python
import re

def faixa_salarial_para_numero(faixa):
    if pd.isna(faixa) or faixa == 'Não informado':
        return np.nan
    elif faixa == 'Menos de R$ 1.000/mês':
        return 500  # Estimativa do ponto médio
    elif faixa == 'Acima de R$ 40.001/mês':
        return 45000 # Estimativa de um valor representativo
    else:
        # Extrai os números da string e remove os separadores de milhar
        numbers = re.findall(r'\d+\.?\d*', faixa.replace('.', ''))
        if len(numbers) == 2:
            lower = float(numbers[0])
            upper = float(numbers[1])
            return (lower + upper) / 2
    return np.nan # Retorna NaN para formatos inesperados

# Aplica a função para criar uma nova coluna numérica
dados['FAIXA_SALARIAL_NUMERICA'] = dados['FAIXA SALARIAL'].apply(faixa_salarial_para_numero)

# Remove valores NaN antes de plotar para evitar erros no boxplot
plt.boxplot(dados['FAIXA_SALARIAL_NUMERICA'].dropna())
plt.title('Boxplot da Faixa Salarial Numérica')
plt.ylabel('Valor Salarial (R$)')
plt.show()
```

```python
Q1 = dados['FAIXA_SALARIAL_NUMERICA'].quantile(0.25)
Q1
```

```python
Q3 = dados['FAIXA_SALARIAL_NUMERICA'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
```

```python
lim_superior = Q3 + 1.5 * IQR
lim_superior
```

```python
lim_inferior = Q1 - 1.5 * IQR
lim_inferior
```

```python
media_salarial = dados['FAIXA_SALARIAL_NUMERICA'].mean()
media_salarial
```

```python
media_salarial = dados['FAIXA_SALARIAL_NUMERICA'].std()
media_salarial
```
